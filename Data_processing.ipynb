## EXT_WTH_EVTS

```python
bl = pd.read_excel(r'CSVs\public_emdat_custom_request_2024-12-01_5dff3569-ccab-4617-ab8f-6864f292db20.xlsx', sheet_name='EM-DAT Data')
```

```python
bl_2 = bl.loc[:, 'Start Year']
```

```python
bl_2.values
```

```python
counter = 0
temp = bl_2.values[0]
output = {}
for x in range(len(bl_2.values)):
    pntr = bl_2.values[x]
    if pntr == temp:
        counter += 1
        print(f'''
              A
              x : {x}
              Counter: {counter}, 
              temp: {temp}, 
              pntr: {pntr}
-----------------------------''')
        if x == len(bl_2.values) -1:
            output.update({temp:counter})
            print(f'''
              B
              x : {x}
              Counter: {counter}, 
              temp: {temp}, 
              pntr: {pntr}
              {output}
-----------------------------''')
        
    else:
        output.update({temp:counter})
        temp = bl_2.values[x]
        counter = 1
        print(f'''
              C
              x : {x}
              Counter: {counter}, 
              temp: {temp}, 
              pntr: {pntr}
              {output}
-----------------------------''')
```

```python
output
```

```python
ext_eve = pd.DataFrame({'FREQ_EXT_EVE': output})
ext_eve.index = [x for x in range(1990, 2025)]
ext_eve = ext_eve.loc[1990:2020, :]
```

```python
ext_eve.tail()
```
